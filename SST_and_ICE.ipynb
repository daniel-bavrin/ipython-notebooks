```python
%pylab inline
# OPeNDAP client: http://pydap.org
from pydap.client import open_url
from netCDF4 import Dataset
import numpy.ma as ma
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
# date to plot. 20130603 is the last non-preliminary by 10.07.2013
date = '20130101' 

# OPeNDAP link (add .html to the end of the url to access with browser)
link = 'http://opendap.solab.rshu.ru:8080/opendap/allData/OISST-AVHRR-AMSR-V2/%s/AVHRR/avhrr-only-v2.%s.nc.gz'%(date[0:4], date)
#local pydap server:
#link = 'http://127.0.0.1:8001/avhrr-only-v2.%s.nc'%(date)

# read data
dataset = open_url(link) 
sst_fill_value = dataset.sst.attributes['_FillValue']
ice_fill_value = dataset.ice.attributes['_FillValue']
sst_add_offset = dataset.sst.attributes['add_offset']
ice_add_offset = dataset.ice.attributes['add_offset']
sst_scale_factor = dataset.sst.attributes['scale_factor']
ice_scale_factor = dataset.ice.attributes['scale_factor']
#read sst and ice and remove singleton dimensions to prepare arrays for pcolor:
sst = np.squeeze(dataset.sst[:])
ice = np.squeeze(dataset.ice[:])
# mask _FillValue values, nulls, scale and shift
sst = ma.masked_where(logical_or(sst == sst_fill_value, sst is None), sst)*sst_scale_factor+sst_add_offset
ice = ma.masked_where(logical_or(ice == ice_fill_value, ice is None), ice)*ice_scale_factor+ice_add_offset

# read lats and lons (representing centers of grid boxes).
lat = dataset.lat[:]
lon = dataset.lon[:]
#shift lats, lons so values represent edges of grid boxes (as pcolor expects):
delon = lon[1]-lon[0]; delat = lat[1]-lat[0]
lon = (lon - 0.5*delon).tolist()
lon.append(lon[-1]+delon)
lon = np.array(lon,np.float64)
lat = (lat - 0.5*delat).tolist()
lat.append(lat[-1]+delat)
lat = np.array(lat,np.float64)

# create a basemap in the EPSG:3413 projection
#proj4: +proj=stere +lat_0=90 +lat_ts=70 +lon_0=-45 +k=1 +x_0=0 +y_0=0 +ellps=WGS84 +datum=WGS84 +units=m +no_defs 
fig = plt.figure(figsize=(15,15))
epsg = 3413; width = 18000.e3; height = 18000.e3
m=Basemap(epsg=epsg,resolution='c',width=width,height=height)

# default 'blue marble' image.
m.arcgisimage()
#m.fillcontinents(color='coral',lake_color='aqua')
m.drawmeridians(np.arange(-180,180,30),labels=[0,0,0,1],color='y')
m.drawparallels(np.arange(0,80,10),labels=[1,0,0,0],color='y')
m.drawcoastlines(linewidth=0.5,color='0.5')

# plot data
# compute map projection coordinates of grid.
x, y = m(*np.meshgrid(lon, lat))

#plot sst, then ice with pcolor
im1 = m.pcolor(x,y,sst,shading='flat',cmap=plt.cm.jet)
im2 = m.pcolor(x,y,ice,shading='flat',cmap=plt.cm.gist_gray)

cb = m.colorbar(im1,"bottom", size="5%", pad="2%")

plt.title('Sea surface temperature for date: %s'%date)
plt.show()
```
